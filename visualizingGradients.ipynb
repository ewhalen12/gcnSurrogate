```python
# generate graph perturbations
graph = allGraphs[i]
r = 0.01
nTheta = 10
thetas = np.linspace(0,2*np.pi,10,endpoint=False)
eps = np.vstack((r*np.cos(thetas),-r*np.sin(thetas))).T
eps.shape
perturbedGraphs = [graph]
for i in range(nTheta):
    g = graph.clone()
    g.pos[0] = g.pos[0] + eps[i,:]
    perturbedGraphs.append(g)
    
```
